# Delete two columns
I deleted two columns because it was not needed

```python
df.drop(
    labels=["Educational level","school system"],
    axis=1,
    inplace=True 
)
df
```

# Edit column names 

Editing some wrong names because the data is translated from Arabic to English

```python
df.rename(columns={"Educational": "University" ,"Detailed area":"Major"}, inplace=True)
df.head()
```

```python
df.rename(columns={"the number": "Number_of_graduates" ,"academic year":"Year"}, inplace=True)
df.head()
```

# Edit values names 
Some values for spelling errors

```python
g= df["gender"].apply(lambda x: "Female"if"feminine"in x else x )
df["gender"] = g
```

```python
df
```

```python
g= df["Major"].apply(lambda x: "Administration and management"if"Management and management"in x else x )
df["Major"] = g
```

```python
df
```

```python
g= df["Major"].apply(lambda x: "Teacher training"if"Training pre-school teachers"in x else x )
df["Major"] = g
```

```python
df
```

```python
g= df["Administrative Region"].apply(lambda x: "AL Baha" if "Patio"in x else x )
df["Administrative Region"] = g
```

```python
df
```

```python
g= df["Administrative Region"].apply(lambda x: "Asir"if "difficult"in x else x )
df["Administrative Region"] = g
```

```python
df
```

# Edit dtypes

```python
df.dtypes
```

```python
g= df["Number_of_graduates"].apply(lambda x: 15 if "15th" in x else x )
df["Number_of_graduates"] = g
```

```python
df = df.astype({"Number_of_graduates": int })
```

```python
df.dtypes
```

```python
df["Year"] = pd.to_datetime(df["Year"])
df.dtypes
```
